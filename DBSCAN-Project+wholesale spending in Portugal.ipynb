## DBSCAN

**TASK: Since the values of the features are in different orders of magnitude, let's scale the data. Use StandardScaler to scale the data.**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
Scaled_X = scaler.fit_transform(df)
```

```python
Scaled_X
```

#### **TASK: Use DBSCAN and a for loop to create a variety of models testing different epsilon values. Set min_samples equal to 2 times the number of features. During the loop, keep track of and log the percentage of points that are outliers.**

```python
from sklearn.cluster import DBSCAN
```

```python
outlier_percent = []

for eps in np.linspace(.001,3,50):
    
    # create the model
    dbscan = DBSCAN(eps=eps,min_samples=2*(Scaled_X.shape[1]))
    dbscan.fit(Scaled_X)
    
    #outlier
    perc_of_outliers = 100 * np.sum(dbscan.labels_ == -1)/len(dbscan.labels_)
    outlier_percent.append(perc_of_outliers)
```

**TASK: Create a line plot of the percentage of outlier points versus the epsilon value choice.**

```python
sns.lineplot(y=outlier_percent,x=np.linspace(.001,3,50))
plt.xlabel('Epsilon Value')
plt.ylabel('Percentage of points classified as Outliers');
```

## DBSCAN with Chosen Epsilon

**TASK: Based on the plot created in the previous task, retrain a DBSCAN model with a reasonable epsilon value. Note: For reference, the solutions use eps=2.**

```python
dbscan = DBSCAN(eps=2)
dbscan.fit(Scaled_X)
```

```python
labels = dbscan.labels_
```
